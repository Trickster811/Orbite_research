```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Load datasets (update the paths if needed)
health_df = pd.read_csv("./wearable/health_wearables.csv")
display(health_df)
ids_df = pd.read_csv("./ids/ids_surveillance.csv")
display(ids_df)


# Plot function
def plot_connectivity_distribution(df, title_prefix):
    plt.figure(figsize=(8, 4))
    sns.histplot(df['Connectivity (Mbps)'], kde=True, bins=10)
    plt.title(f"{title_prefix}: Connectivity (Mbps) Distribution")
    plt.xlabel("Connectivity (Mbps)")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

# Plot for both datasets
plot_connectivity_distribution(health_df, "Health Wearables (Capped)")
plot_connectivity_distribution(ids_df, "IDS Surveillance (Capped)")
```
